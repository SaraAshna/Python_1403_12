```python
#5
def firstUChar(clip):  
    clip = clip.split(' ')
    word_count = {}
    for char in clip:
        word_count.setdefault(char,0)
        if char not in word_count:
            word_count.append(char)
        else:
            word_count[char]+=1  
    for k,v in word_count.items():
        if v==1:
            print(k)
            break
    else:
        print(-1)

firstUChar('apple apple nine ten fruit ten nine apple mango banana banana mango apple watermelon fruit basket')
```

```python
#6
def can_be_mirror(str_):
    count_letters = {}
    count_odd = 0
    for char in str_:
        if char in count_letters:
            count_letters[char] += 1
        else:
            count_letters[char] = 1

    for char in count_letters:
        if count_letters[char] % 2 == 1:
            count_odd += 1
    if count_odd <=1:
        return True
    else:
        for count in count_letters.values():
            if count >= 2:
                return True 

        return False  
```

```python
can_be_mirror("rahkar")
```
