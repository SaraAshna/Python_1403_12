```python
#1
def get_words(file_name):
    words = set()
    with open(file_name,'r') as f1:
        for lines in f1:
            for word in lines.split():
               only_word = word.strip().lower().strip('. , ? ()')
               if only_word:
                   words.add(word)
        return words

def find_M(file1,file2):
    words1 = get_words(file1)
    words2 = get_words(file2)
    return len(words1.intersection(words2))


find_M('t1.txt','t2.txt')

```

```python
#2
def copy_file(f1,f2):
    with open(f1,'r') as f:
        c = f.read()
    with open (f2,'a') as new:
        new.write(c)

copy_file('t1.txt','t4.txt')
```

```python
#3
def remove_empty_lines(file_name):
    new_text = ''
    with open(file_name,'r') as f:
        for line in f:
            if line.strip():
                new_text += line
        with open(file_name,'w') as new:
            new.write(new_text)

remove_empty_lines('t5.txt')
```

```python
#4
def search_word(file_name,term):
    words_count = {}
    with open(file_name,'r') as f:
        for lines in f:
            for word in lines.split():
                clean_word = word.strip()
                if clean_word:
                    if clean_word == term:
                        words_count.setdefault(clean_word,0)
                        words_count[clean_word] +=1
        return words_count


search_word('t1.txt','Male')
```

```python
#5
def compare_files(file_1,file_2):
    with open(file_1,'r') as f1, open(file_2,'r') as f2:
        if f1.read() == f2.read():
            return True
        return False

compare_files('t1.txt','t2.txt')
```

```python
#6
def find_lines_with_word(file_name,term):
    with open(file_name,'r') as f:
        all_lines = []
        lines = f.readlines()
        for line in lines:
            if term in line:
                all_lines.append(line)
        return all_lines
find_lines_with_word('t1.txt','Female')
```

```python
#7
def get_words(file_name):
    words = set()
    with open(file_name,'r') as f1:
        for lines in f1:
            for word in lines.split():
               only_word = word.strip().lower().strip('. , ? ()')
               if only_word:
                   words.add(word)
        return words

def find_intersection(file1,file2):
    words1 = get_words(file1)
    words2 = get_words(file2)
    return words1.intersection(words2)


def file_similarity_difference(file1,file2):
    with open(file1,'r') as f1, open(file2,'r') as f2:
        w1 = get_words(file1)
        w2= get_words(file2)
        intersection =len(find_intersection(file1,file2))
        union = len(w1)+len(w2)-intersection
        similarity = intersection/union
        dissimilarity= 1 - similarity
        return ('similarity:',similarity*100 , 'dissimilarity:',dissimilarity*100, len(w1),len(w2))

file_similarity_difference('t1.txt','t2.txt')
```

```python
#8
def file_statistics(file_name):
    file_info = {'lines_count':0,'words_count':0,'chars':0}
    with open(file_name,'r') as f:
        for line in f:
            if line.strip(): 
                file_info['lines_count'] +=1
                for words in line.split(','):
                    for word in words.strip().split():
                        if word.isalpha():
                            file_info['words_count']+= 1
                    for char in words:
                        file_info['chars']+=len(char)
        return file_info

file_statistics('t1.txt')
```
