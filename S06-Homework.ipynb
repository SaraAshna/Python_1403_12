d1 = {'a': 100, 'b': 200, 'c':300}

d2 = {'a': 300, 'b': 200, 'd':400}

خروجی مورد انتظار ({'a': 400, 'b': 400, 'd': 400, 'c': 300})

```python
d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}
d3 = {}
for j in d1.keys():
    if j in d2:
      d3[j] = d1[j] + d2[j]
    else:
       d3[j] = d1[j]
for j in d2.keys():
    if j in d1:
      d3[j] = d1[j] + d2[j]
    else:
       d3[j] = d2[j]   
print(d3)

```

یک لیست از تاپلهای حاوی نام و سن افراد را درست کنید و خروجی زیر را تولید کنید
یک ست از سن های منحصربه فرد.
یک دیکشنری که تعداد افراد در هر سن را نشان میدهد.
یک لیست از نامها به ترتیب سن صعودی

ورودی مورد انتظار


people = [("Ali", 25), ("Mina", 30), ("Reza", 25), ("Sara", 30)]

خروجی مورد انتظار

سن های منحصربه فرد
{25, 30}

تعداد افراد در هر سن 
{25: 2, 30: 2}

نامها بر اساس سن
["Ali", "Reza", "Mina", "Sara"]

```python
people = [("Ali", 25), ("Mina", 30), ("Reza", 25), ("Sara", 30)]
result = {}
ages = []
for k,v in people:
    result.setdefault(v,0)
    ages.append(v)
    if v in result:
        result[v]+=1
print(set(ages))
print(result)
names = []
d = dict (people)
sorted_d = sorted(d.values())
for x in sorted_d:
    for i,item in d.items():
        if x ==item and i not in names :
            names.append(i)
print(names)
```

لیستی از کلمات را دریافت کنید و یک دیکشنری بسازید که کلیدهای آن طول کلمات و مقدار آن لیست کلمات با آن طول باشد

ورودی نمونه

words = ["apple", "cat", "dog", "book", "python"]

خروجی مورد انتظار

{5: ['apple'], 3: ['cat', 'dog'], 4: ['book'], 6: ['python']}

 

```python
words = ["apple", "cat", "dog", "book", "python"]

my_dict = {}

for word in words:
    length = len(word)
    if length not in my_dict:
        my_dict[length] = []
    my_dict[length].append(word)

print(my_dict)
```
