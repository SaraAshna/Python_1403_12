```python
class IceCream():
    STOCK = 200000
    Sold_amount = 0
    PROFIT = 0
    def __init__(self, weight, price):
        self.weight = weight
        self.price = price

    def sell(self, count):
        amount = count*self.weight
        IceCream.PROFIT += (self.price*count)
        if amount<= IceCream.STOCK:
            IceCream.STOCK -= amount
            IceCream.Sold_amount +=amount
            return amount, IceCream.STOCK
        return 'finished'
    
    def remained():
        return (IceCream.STOCK)    
        
    def refill_stock(self):
        IceCream.STOCK = IceCream.STOCK + IceCream.Sold_amount
        return IceCream.STOCK
    
    def profit_calculator():
        return (IceCream.PROFIT*0.2)
    

class MUSICAL_ICE(IceCream):
    
    def __init__(self, weight, price,song):
        super().__init__(weight, price)
        self.song= song
    
    def play_music(self):
        print(f'{self.song} is playing')


```

```python
fruti = IceCream(55,7)
choco = IceCream(100,10)
icy_pop = IceCream(40,3)
musical_ice = MUSICAL_ICE(400,15,'Moon River')
```

```python
musical_ice.sell(5)
```

```python
musical_ice.play_music()
```
