```python
#4
def distance(x1,y1,x2,y2):
    d = ((x2-x1)**2+(y2-y1)**2)**0.5
    return d
distance(-2,1,2,4)
```

```python
#5
def collinear(x1,y1,x2,y2,x3,y3):
    if (y2-y1)/(x2-x1) == (y3-y2)/(x3-x2):
        return ('on the same line')
    else:
        return ('Not on the same line')

collinear(4, 4,-2, 6,1, 5)
```

```python
#6
def triangle(x1,y1,x2,y2,x3,y3):

    def distance(ax,ay,bx,by):
        return(((ax-bx)**2+(ay-by)**2)**0.5)
    
    AB = distance(x1,y1,x2,y2)
    BC = distance(x2,y2,x3,y3)
    AC = distance(x1,y1,x3,y3)

    area = 0.5 * abs(x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2))
    if area == 0:
        return('not triangle')
    else:
        if AB == AC ==BC :
            return ("Equilateral triangle")
        if AB == AC or AC == BC or BC == AB:
            return ("Isoceles triangle")
        else:
            return ('Scalene triangle')
        
print(triangle(0, 0, 3, 0, 1, 2))         
print(triangle(0, 0, 1, 1, 2, 2))  
    
```

```python
#7
def ori_dis(*args, origin = (0,0)):
        result = []
        for i in range(0,len(args),2):
              x,y = args[i],args[i+1]
              distance = ((x-origin[0])**2+(y-origin[1])**2)**0.5
              result.append(distance)

        return sorted(result)

ori_dis(25,4,68,4,59,87,74,7,58,22,14,1)
```
